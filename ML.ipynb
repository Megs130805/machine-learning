```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report

# Load Excel file (make sure you've uploaded it)
df = pd.read_excel('plant_disease_dataset.xlsx')  # Use correct file name

# Create 2 random numeric features for simulation
np.random.seed(42)
df['feature1'] = np.random.uniform(0, 10, size=len(df))
df['feature2'] = np.random.uniform(0, 10, size=len(df))

# Define features and target
X = df[['feature1', 'feature2']]
y = df['label']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train kNN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Predictions
y_pred_train = knn.predict(X_train)
y_pred_test = knn.predict(X_test)

# Confusion Matrices
print("=== Train Confusion Matrix ===")
print(confusion_matrix(y_train, y_pred_train))

print("\n=== Test Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred_test))

# Classification Reports
print("\n=== Train Classification Report ===")
print(classification_report(y_train, y_pred_train))

print("\n=== Test Classification Report ===")
print(classification_report(y_test, y_pred_test))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
import numpy as np

# Simulated continuous target: let's say this is some 'severity score'
# In real use, this would come from the dataset itself
np.random.seed(42)
severity_score = np.random.uniform(0, 100, len(X))  # Fake target values

# Train-test split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, severity_score, test_size=0.3, random_state=42)

# Train regression model
reg = LinearRegression()
reg.fit(X_train_reg, y_train_reg)

# Predict
y_pred_reg = reg.predict(X_test_reg)

# Metrics
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test_reg, y_pred_reg)
r2 = r2_score(y_test_reg, y_pred_reg)

print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAPE: {mape:.4f}")
print(f"R² Score: {r2:.4f}")

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Load dataset
df = pd.read_excel('plant_disease_dataset.xlsx')  # Make sure file is uploaded

# Step 2: Create two dummy numeric features (if you haven't already)
np.random.seed(42)
df['feature1'] = np.random.uniform(1, 10, len(df))
df['feature2'] = np.random.uniform(1, 10, len(df))

# Step 3: Plot training data (A3)
X_train = df[['feature1', 'feature2']]
y_train = df['label']

colors = ['blue' if label == 0 else 'red' for label in pd.factorize(y_train)[0]]
plt.figure(figsize=(6, 5))
plt.scatter(X_train['feature1'], X_train['feature2'], c=colors)
plt.title("A6: Scatter Plot of Project Data")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.grid(True)
plt.show()

# Step 4: Create test set of 10,000 points (A4)
xx, yy = np.meshgrid(np.arange(0, 10, 0.1), np.arange(0, 10, 0.1))
test_points = np.c_[xx.ravel(), yy.ravel()]

# Encode class labels
y_encoded = pd.factorize(y_train)[0]

# Step 5: Train kNN and predict test data
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_encoded)
predicted = knn.predict(test_points)

# Step 6: Plot classification result (A4)
colors_pred = ['blue' if c == 0 else 'red' for c in predicted]

plt.figure(figsize=(6, 5))
plt.scatter(test_points[:, 0], test_points[:, 1], c=colors_pred, s=1, alpha=0.3)
plt.title("A6: kNN Classification (k=3)")
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.grid(True)
plt.show()

# Step 7: Repeat with different k values (A5-style)
for k in [1, 3, 5, 7, 11]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_encoded)
    predicted = knn.predict(test_points)

    colors_pred = ['blue' if c == 0 else 'red' for c in predicted]

    plt.figure(figsize=(6, 5))
    plt.scatter(test_points[:, 0], test_points[:, 1], c=colors_pred, s=1, alpha=0.3)
    plt.title(f"A6: kNN Classification (k={k})")
    plt.xlabel("feature1")
    plt.ylabel("feature2")
    plt.grid(True)
    plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': list(range(1, 20))}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid.fit(X_train, y_train)

print("Best k value:", grid.best_params_)
print("Best score:", grid.best_score_)
```
